```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## **normal distribution**

```python

s = [10, 100, 1000]
def Normal(sizes):
    th = np.linspace(-4, 4, 400)

    fig, ax = plt.subplots(1, 3, figsize=(18,5))
    fig.suptitle('Normal distribution N(0,1)', fontsize=16)

    for i, n in enumerate(sizes):
        d = stats.norm(loc=0, scale=1).rvs(size=n)

        ax[i].set_title(f'n = {n}')
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('density')

        bins = int(1 + np.log2(n))  # формула Стерджеса

        ax[i].hist(d, bins=bins, density=True, edgecolor='black', color='blue', alpha=0.75)
        ax[i].plot(th, stats.norm.pdf(th, 0, 1), color ='red', linewidth=2)

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

Normal(s)
```

```python
s = [10, 100, 1000]
def Koshi(s):
    th =np.arange(-15, 15, 0.001)

    fig, ax = plt.subplots(1, 3, figsize=(16,4))
    fig.suptitle('Cauchy distribution C(0,1)', fontsize=16)
    
    for i,n in enumerate(s):
        practice = stats.cauchy(loc=0, scale=1).rvs(size=n)

        ax[i].set_title(f'n = {n}')
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('density')

        render_range = (-15, 15) 
    
        ax[i].hist(practice, bins=int(np.sqrt(n)), range=render_range, 
               density=True, edgecolor='black', color='blue', alpha=0.75)
        ax[i].plot(th, stats.cauchy.pdf(th, 0, 1), color='red', linewidth=2)    
        ax[i].set_xlim(render_range)

    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

Koshi(s)
```

```python
size = [10, 100, 1000]
def Uniform(size):
    th = np.arange(-2, 2, 0.001)
    
    fig, ax = plt.subplots(1, 3, figsize=(16,4))
    fig.suptitle('Uniform distribution', fontsize=16)
    a = -np.sqrt(3)
    b = -(a)
    for i, n in enumerate(size):
        
        practice = stats.uniform(loc= a, scale=b-a).rvs(size = n)


        ax[i].set_title(f'n = {n}')
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('density')

        bins = int(1 + np.log2(n))  # формула Стерджеса
        print(bins)

        ax[i].set_title(f'n = {n}')
        ax[i].hist(practice, bins=int(np.sqrt(n)), density=True, edgecolor='black', color='blue', alpha=0.75)
        ax[i].plot(th, stats.uniform.pdf(th,loc = a,scale =b-a), color ='red', linewidth=2)

        # plt.hist(practice, bins=20, density=True, edgecolor='black', color='blue', alpha=0.75)
        # plt.plot(th, stats.uniform.pdf(th, loc=-3**0.5, scale=2*3**0.5), color ='red', linewidth=2)
    plt.show()

Uniform(size)
```

```python
size = [10, 100, 1000]
def Laplace(size):
    th = np.arange(-4, 4, 0.001)
    
    fig, ax = plt.subplots(1, 3, figsize=(16,4))
    fig.suptitle('Laplace distribution', fontsize=16)
    a = 0
    b = 1/np.sqrt(2)

    for i, n in enumerate(size):
        practice = stats.laplace(loc = a, scale=b).rvs(size=n)
        ax[i].set_title(f'n = {n}')
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('density')

        bins = int(np.sqrt(n))  
        print(bins)

        ax[i].set_title(f'n = {n}')
        ax[i].hist(practice, bins=int(np.sqrt(n)), density=True, edgecolor='black', color='blue', alpha=0.75)
        ax[i].plot(th, stats.laplace.pdf(th,loc = a,scale =b-a), color ='red', linewidth=2)

    plt.show()

Laplace(size)
```

```python
size = [10, 100, 1000]
def Puasson(size):
    th = np.arange(stats.poisson.ppf(0.001, mu=5), stats.poisson.ppf(0.999, mu=5))
    
    fig, ax = plt.subplots(1, 3, figsize=(16,4))
    fig.suptitle('Poisson distribution', fontsize=16)

    for i, n in enumerate(size):
        practice = stats.poisson(mu=5).rvs(size=n) 
      
        ax[i].set_title(f'n = {n}')
        ax[i].set_xlabel('x')
        ax[i].set_ylabel('density')

        bins = np.arange(-0.5, max(practice)+1.5, 1)
        print(bins)

        ax[i].set_title(f'n = {n}')
        ax[i].hist(practice, bins=bins, density=True, edgecolor='black', color='blue', alpha=0.75)
        ax[i].plot(th, stats.poisson.pmf(th, mu=5), color ='red', linewidth=2)

    plt.show()

Puasson(size)
```
