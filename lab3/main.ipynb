```python
import matplotlib.pyplot as plt
import scipy.stats as scs
import pandas as pd
import numpy as np
import seaborn as sns
```

```python

distributions = {
    "Normal": lambda n: scs.norm.rvs(0, 1, size=n),
    "Cauchy": lambda n: scs.cauchy.rvs(0, 1, size=n),
    "Laplace": lambda n: scs.laplace.rvs(0, scale=1/np.sqrt(2), size=n),
    "Poisson": lambda n: scs.poisson.rvs(5, size=n),
    "Uniform": lambda n: scs.uniform.rvs(-np.sqrt(3), 2*np.sqrt(3), size=n)
}
```

```python
def calculate_emmisions(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    n = len(x)
    x1 = q1 - 1.5*(q3-q1)
    x2 = q3 + 1.5*(q3-q1)

    count = 0
    for i in range(0, n):
        if(x[i] < x1  or x[i] > x2):
            count += 1
        
    print(count)
    return count
```

```python

def tukey_outliers_fraction(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = np.sum((x < lower) | (x > upper))
    return outliers / len(x)


num_experiments = 1000

results = {}

for name, dist in distributions.items():
    results[name] = {}

    for n in n_val:
        fractions = []

        for _ in range(num_experiments):
            sample = dist(n)
            frac = tukey_outliers_fraction(sample)
            fractions.append(frac)

        results[name][n] = np.mean(fractions)

print("\nСредняя доля выбросов (1000 повторений):\n")

for name in results:
    print(name)
    for n in results[name]:
        print(f"  n={n}: {results[name][n]:.4f}")
    print()
```
